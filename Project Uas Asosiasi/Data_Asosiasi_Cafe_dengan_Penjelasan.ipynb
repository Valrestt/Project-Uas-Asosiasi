## Penjelasan Cell

**Tujuan**

Cell ini merupakan bagian dari tahapan analisis. Jalankan cell untuk memperoleh hasil sesuai proses sebelumnya, kemudian interpretasikan output yang dihasilkan.

**Fungsi**

Menjalankan tahapan analisis sesuai output yang dihasilkan pada cell kode di bawah ini.

`````python
# =========================================================
# Membuat Data Transaksi dari Kolom association_items
# =========================================================

# Mengubah teks pada kolom association_items menjadi list item per transaksi.
# Contoh: "gender_M, age_Muda, offer_bogo" menjadi ['gender_M', 'age_Muda', 'offer_bogo']
transactions = []

for item_text in df_clean['association_items']:
    items = [item.strip() for item in str(item_text).split(',') if item.strip() != '']
    transactions.append(items)

print("✅ Data berhasil diubah menjadi format daftar transaksi.")
print(f"Jumlah transaksi: {len(transactions):,}")

print("\n🔍 Contoh 5 transaksi pertama:")
for i, trx in enumerate(transactions[:5], start=1):
    print(f"Transaksi {i}: {trx}")
`````

## Penjelasan Cell

**Tujuan**

Cell ini merupakan bagian dari tahapan analisis. Jalankan cell untuk memperoleh hasil sesuai proses sebelumnya, kemudian interpretasikan output yang dihasilkan.

**Fungsi**

Menjalankan tahapan analisis sesuai output yang dihasilkan pada cell kode di bawah ini.

`````python
# =========================================================
# Encoding Data dengan TransactionEncoder
# =========================================================

# TransactionEncoder mengubah list transaksi menjadi format one-hot/biner True-False.
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

# Membuat dataframe encoded
df_encoded = pd.DataFrame(te_array, columns=te.columns_)

print("✅ Encoding selesai!")
print(f"Shape data encoded: {df_encoded.shape[0]} transaksi x {df_encoded.shape[1]} item")

# Menampilkan data encoded dalam bentuk 0/1 agar lebih mudah dibaca
df_encoded.astype(int).head()
`````

`````python
# =========================================================
# Fungsi Bantuan untuk Merapikan Rules
# =========================================================

def frozenset_to_text(x):
    """Mengubah frozenset menjadi teks agar mudah dibaca."""
    return ', '.join(list(x))


def rapikan_rules(rules_df):
    """Menambahkan kolom teks dan mengurutkan rules."""
    rules_df = rules_df.copy()

    # Tetap buat kolom teks walaupun rules kosong agar cell berikutnya tidak error.
    if 'antecedents' in rules_df.columns:
        rules_df['antecedents_str'] = rules_df['antecedents'].apply(frozenset_to_text)
        rules_df['antecedent_len'] = rules_df['antecedents'].apply(lambda x: len(x))
    else:
        rules_df['antecedents_str'] = []
        rules_df['antecedent_len'] = []

    if 'consequents' in rules_df.columns:
        rules_df['consequents_str'] = rules_df['consequents'].apply(frozenset_to_text)
        rules_df['consequent_len'] = rules_df['consequents'].apply(lambda x: len(x))
    else:
        rules_df['consequents_str'] = []
        rules_df['consequent_len'] = []

    if not rules_df.empty and {'lift', 'confidence', 'support'}.issubset(rules_df.columns):
        rules_df = rules_df.sort_values(['lift', 'confidence', 'support'], ascending=False)

    return rules_df


def tampilkan_rules(rules_df, n=10, judul='Association Rules'):
    """Menampilkan rules dengan kolom penting."""
    print(f"📋 {judul}")
    print("-" * 80)
    if rules_df.empty:
        print("Tidak ada rules yang memenuhi parameter.")
        return pd.DataFrame()
    kolom_tampil = [
        'antecedents_str', 'consequents_str', 'support', 'confidence',
        'lift', 'leverage', 'conviction'
    ]
    return rules_df[kolom_tampil].head(n)
`````

## 🧪 Penjelasan Training dan Testing pada Association Rule Mining

Association Rule Mining tidak memiliki proses training/testing seperti klasifikasi, karena tidak ada target/class yang diprediksi. Namun, untuk memenuhi ketentuan UAS, data tetap dibagi menjadi:

1. **Data Training**: digunakan untuk membentuk frequent itemsets dan association rules.
2. **Data Testing**: digunakan untuk mengecek apakah rules yang terbentuk pada data training masih muncul dan relevan pada data testing.

Evaluasi pada data testing dilakukan dengan menghitung:

- **Antecedent Count**: jumlah transaksi testing yang mengandung antecedent.
- **Both Count**: jumlah transaksi testing yang mengandung antecedent dan consequent sekaligus.
- **Test Confidence**: `both_count / antecedent_count`.
- **Coverage**: proporsi transaksi testing yang tercakup oleh antecedent rule.

## Penjelasan Cell

**Tujuan**

Cell ini merupakan bagian dari tahapan analisis. Jalankan cell untuk memperoleh hasil sesuai proses sebelumnya, kemudian interpretasikan output yang dihasilkan.

**Fungsi**

Menjalankan tahapan analisis sesuai output yang dihasilkan pada cell kode di bawah ini.

`````python
# =========================================================
# Evaluasi Rules pada Data Testing
# =========================================================

def evaluasi_rules_testing(rules_df, test_df, top_n=50):
    """Mengevaluasi rules pada data testing berdasarkan test confidence dan coverage."""
    hasil = []
    test_bool = test_df.astype(bool)

    # Ambil rules terbaik berdasarkan lift, confidence, dan support
    rules_eval = rules_df.sort_values(['lift', 'confidence', 'support'], ascending=False).head(top_n)

    for _, row in rules_eval.iterrows():
        antecedents = list(row['antecedents'])
        consequents = list(row['consequents'])

        # Mengecek apakah semua item antecedent dan consequent tersedia pada data testing
        if not set(antecedents).issubset(test_bool.columns) or not set(consequents).issubset(test_bool.columns):
            continue

        mask_antecedent = test_bool[antecedents].all(axis=1)
        antecedent_count = mask_antecedent.sum()

        if antecedent_count == 0:
            test_confidence = np.nan
            both_count = 0
        else:
            mask_consequent = test_bool[consequents].all(axis=1)
            both_count = (mask_antecedent & mask_consequent).sum()
            test_confidence = both_count / antecedent_count

        coverage = antecedent_count / len(test_bool)

        hasil.append({
            'antecedents': row['antecedents_str'],
            'consequents': row['consequents_str'],
            'train_support': row['support'],
            'train_confidence': row['confidence'],
            'train_lift': row['lift'],
            'antecedent_count_test': antecedent_count,
            'both_count_test': both_count,
            'test_confidence': test_confidence,
            'coverage': coverage
        })

    return pd.DataFrame(hasil)

# Evaluasi menggunakan rules FP-Growth karena FP-Growth umumnya lebih efisien.
eval_test_fpgrowth = evaluasi_rules_testing(rules_fpgrowth_filtered, test_encoded, top_n=50)

print("✅ Evaluasi data testing selesai!")
eval_test_fpgrowth.head(15)
`````
